```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

# Matplotlib basics

```python
x = np.linspace(0,5,11)
x

y = x ** 2
y
```

```python
plt.plot (x ,y)
plt.title("first graph")
plt.xlabel("x label")
plt.ylabel("y label")
```

# subplotes in matplotlib

```python
plt.subplot(2,2,1)
plt.plot(y**2 ,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2 ,x**2)
```

# object oriented way 

matplotlib allows the aspect ratio , dpi (dots per inch)[pixel per inch] and figure size to be specified when the figure object is created#
-use figsize and dpi keyword arguments 

figsize is a tuple of the width and height of the figure in inches dpi 

```python
fig = plt.figure(figsize=(20,8) ,dpi = 100)
axis1 = fig.add_axes([0.1,0.1 ,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.1 ,0.3,0.3])
axis2.plot(x,y)
```

# saving plots 

```python
fig = plt.figure()
ax= fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
