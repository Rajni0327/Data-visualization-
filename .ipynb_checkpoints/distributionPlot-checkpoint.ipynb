# it is used for the distribution of quantitative data 

plots are :
1. displot / histplot
2. jointplot
3. pairplot
4. rugplot


```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['size'].unique()
```

# displot / histplot

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'] ,bins = 20 ,kde = True)  #default bin = 20
plt.subplot(1,2,2)
sns.histplot(df["tip"] ,bins = 20, kde = True)
```

# jointplot

```python
sns.jointplot(x = 'total_bill' , y = "tip" ,data = df, kind='scatter')
```

```python
sns.jointplot(x = 'total_bill' , y = "tip" ,data = df, kind='kde')
```

```python
sns.jointplot(x = 'total_bill' , y = "tip" ,data = df, kind='hex')
```

```python
sns.jointplot(x = 'total_bill' , y = "tip" ,data = df, kind='reg')
```

# pairplot

```python
#for analyzing
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df , hue ='sex')   # for differentiating btw male and female
```

```python
sns.pairplot(df , hue ='size',palette='rainbow')
```

# rugplot

```python
sns.rugplot(df)
```
